#### 'strength' 컬럼추가 : alcohol을 기준으로 10도가 넘으면 'strong' 아니면 'weak'으로 표시 (상위 30개만 출력)

```python
def is_strong(alcohol):
    if alcohol < 10 :
        return 'weak'
    else :
        return 'strong'
```

```python
data['strength'] = data['alcohol'].apply(is_strong)
data.head(30)
```

```python
df_y = df_y[df_y.quality >= 7]
```

```python
len(df_y)
```

#### df_x와 df_y를 inner join시키기 (quality를 기준으로)

```python
pd.merge(df_x, df_y, how = 'inner', on = 'quality')
```

#### df_x에서 df_y를 left join시키기 (quality를 기준으로)

```python
pd.merge(df_x, df_y, how = 'left', on = 'quality')
```

#### birthDate 컬럼으로 age 컬럼 추가하기

##### 1) birthDate 컬럼의 자료형을 object에서 datetime으로 전환

```python
data['birthDate'] = pd.to_datetime(data['birthDate'])
```

##### 2) age 컬럼 생성, apply 및 lambda 함수를 써서 나이 계산

```python
from datetime import datetime
today = datetime.now().date()

data['age'] = data['birthDate'].apply(lambda x : today.year - x.year)

data.info()
```

##### 3) age 결측치를 선수 평균 나이로 대체

```python
mean = int(data['age'].mean()) # average age of Players
data['age'] = data['age'].fillna(mean)
data['age'] = data['age'].astype('int')
data.info()
```

#### height 컬럼(feet-inch 형식)으로 heightCm 컬럼 추가하기

##### 1) heightCm 컬럼 생성, apply 및 lambda 함수를 사용해서 height 컬럼의 feet과 inch를 cm로 변환 ('-'를 기준으로 split해서 각각을 cm로 변환해 더하기)



```python
data['heightCm'] = data['height'].apply(lambda x : x.split('-'))
data['heightCm'] = data['heightCm'].apply(lambda y : int(y[0])*30.48 + int(y[1])*2.54)
data.head(30)
```
